```python
import pandas as pd
import numpy as np

df= pd.read_csv('G:\\Python coding\\Metropolia\\ML Projects\\Dataset\\cancerdata.csv')



```

```python
df.drop('Unnamed: 32',inplace= True, axis=1)
df.head()
```

```python
df['diagnosis'].unique



df['diagnosis'].value_counts()
```

```python
# Use labelencoder o label encode values in Diagnosis column

from sklearn.preprocessing import LabelEncoder

labelecoder=LabelEncoder()
labelecoder.fit_transform(df.iloc[:,1].values)
```

```python
#split data into indepedent & dependent features

X=df.iloc[:,2:].values
y=df.iloc[:,1].values
```

```python
# Task 4: Split data as Train test

from sklearn.linear_model import LinearRegression
from sklearn.model_selection  import train_test_split

X_train,X_test,y_train,y_test=train_test_split(X,y,random_state=100,train_size=0.75)
```

```python
# perform feature scaling

from sklearn.preprocessing import StandardScaler

sc=StandardScaler()
X_train=sc.fit_transform(X_train)
X_test=sc.fit_transform(X_test)
```

```python
# build logistic regression
from sklearn.linear_model import LogisticRegression

lr=LogisticRegression()
#y_train = y_train.astype(int)		#Y_train to numeric

#y_test = y_test.astype(int)		        #Y_test to numeric
mapping = {'M': 1, 'B': 0}
y_train = np.array([mapping[value] for value in y_train])
y_test =  np.array([mapping[value] for value in y_test])
lr.fit(X_train,y_train)
lr.get_params()
```

```python
# do the predections
pred=lr.predict(X_test)
```

```python
#calculate accuracy
from sklearn.metrics import confusion_matrix, accuracy_score
import seaborn as sn
cm=confusion_matrix(y_test,pred)
sn.heatmap(cm, annot= True)

score=accuracy_score(y_test,pred)
print(score)
```
