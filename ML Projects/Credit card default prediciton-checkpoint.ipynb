```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_excel('G:\\Python coding\\Metropolia\\ML Projects\\Dataset\\default of credit card clients.xls')
df.head()
```

```python
df.shape
df.info
df.describe()
```

```python
print(df.isnull().sum())
```

```python
df = df.apply(pd.to_numeric, errors='coerce')  # Convert all object columns to float
df.fillna(0, inplace=True)  # Replace NaN values with 0 (optional)

df.info()
```

```python
# Lets take X,y values

X=df.iloc[:,1:24].values
y=df.iloc[:,24:25].values
```

```python
from sklearn.model_selection  import train_test_split

X_train,X_test,y_train,y_test=train_test_split(X,y,random_state=42,train_size=0.60)
```

```python
# now lets scale the independent variables
from sklearn.preprocessing import StandardScaler

sc=StandardScaler()
X_train=sc.fit_transform(X_train)
X_test=sc.fit_transform(X_test)
```

```python
from sklearn.linear_model import LogisticRegression

lr=LogisticRegression()
lr.fit(X_train,y_train)
```

```python
y_pred = lr.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
accuracy = accuracy_score(y_test, y_pred)
conf_matrix = confusion_matrix(y_test, y_pred)
class_report = classification_report(y_test, y_pred)
```

```python
print('accuracy:',accuracy)   # We got 81% accuracy with Logistic regression
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(5, 4))
sns.heatmap(conf_matrix, annot=True, fmt="d", cmap="Blues", xticklabels=["No Default", "Default"], yticklabels=["No Default", "Default"])
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title("Confusion Matrix")
plt.show()
```

## model_simple is a one-layer network (ie. ~ logistic regression)

```python

import tensorflow as tf
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense



model = tf.keras.Sequential([tf.keras.layers.Dense(1, activation='sigmoid', input_shape=(X_train.shape[1],))]) # Logistic Regression Layer
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])   # binary_crossentropy is used since is logistic regression

# Train the model for 50 epochs and see Accuracy & loss function
history_simple=model.fit(X_train, y_train, epochs=50, batch_size=32, validation_data=(X_test, y_test), verbose=1)
loss, accuracy = model.evaluate(X_test, y_test)
```

```python
# Train the model for 100 epochs and see Accuracy & loss function
model.fit(X_train, y_train, epochs=100, batch_size=32, validation_data=(X_test, y_test), verbose=1)
loss, accuracy = model.evaluate(X_test, y_test)
```

# Findings:

In traditional machine Learning when Logisitic regresson applied on same data the accuracy is : 81 %

When using nureal netowrk, with single layer , 50 epochs accuracy of validation data is : 81 %

When using nureal netowrk, with single layer , 50 epochs accuracy of validation is : 81 %

I dont see any improvement when data passed through CNN with single layer.

## Model_complex is 5 layer network with enough width

```python
model_fivelayer = tf.keras.Sequential([
    tf.keras.layers.Dense(128, activation='relu', input_shape=(X_train.shape[1],)),  # Layer 1
    tf.keras.layers.BatchNormalization(),  # Normalize activations
    tf.keras.layers.Dropout(0.3),  # Prevent overfitting
    
    tf.keras.layers.Dense(64, activation='relu'),  # Layer 2
    tf.keras.layers.BatchNormalization(),
    tf.keras.layers.Dropout(0.3),

    tf.keras.layers.Dense(32, activation='relu'),  # Layer 3
    tf.keras.layers.BatchNormalization(),
    
    tf.keras.layers.Dense(16, activation='relu'),  # Layer 4
    
    tf.keras.layers.Dense(1, activation='sigmoid')  # Output Layer (Binary Classification)
])

model_fivelayer.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

history_complex=model_fivelayer.fit(X_train, y_train, epochs=50, batch_size=32, validation_data=(X_test, y_test), verbose=1)
loss, accuracy = model_fivelayer.evaluate(X_test, y_test)
```

```python
plt.figure(figsize=(12, 5))
    
# Plot training accuracy
plt.subplot(1, 2, 1)
plt.plot(history_simple.history['accuracy'], label='Simple - Training')
plt.plot(history_complex.history['accuracy'], label='Complex - Training')
plt.title('Training Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()
    
# Plot validation accuracy
plt.subplot(1, 2, 2)
plt.plot(history_simple.history['val_accuracy'], label='Simple - Validation')
plt.plot(history_complex.history['val_accuracy'], label='Complex - Validation')
plt.title('Validation Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()
    
plt.tight_layout()
plt.show()
```

## Findings

As shown in the diagram. i have compared simple layer vs 5 layer dense network with 50 epochs
the accuracy improved in both Training & Validation data, when used CNN with 5 layers dense network.
