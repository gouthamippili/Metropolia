```python
import tensorflow as tf

w=tf.Variable([.2])  # initial weight, it can be anything
b=tf.Variable([-.2]) # initial bias, it can be anything
x=tf.compat.v1.placeholder(tf.float32)  # input 
```

```python
linear_model=w*x + b

init = tf.compat.v1.global_variables_initializer()  # initalizing variables
```

```python
s=tf.compat.v1.Session() # setting up session for computation

s.run(init)
s.run(w)
s.run(b)
```

```python
print("predicted output is")
print(s.run(linear_model,{x:[1,2,3,4]})) # applying this formula w*x + b with initial value of weight and bias
```

```python
y=tf.compat.v1.placeholder(tf.float32)  # y is for holding desired output
squared_deltas=tf.square(linear_model -y)
loss = tf.reduce_sum(squared_deltas) # reduce_sum is used to sum all the squared errors and give us rmse
```

```python
print("loss or Error is ")
print(s.run(loss, {x:[1,2,3,4], y:[0,-1,-2,-3]}))   #given input and output values here because we have used placeholder
```

```python
import tensorflow as tf

w=tf.compat.v1.Variable([.2])  # initial weight, it can be anything
b=tf.compat.v1.Variable([-.2]) # initial bias, it can be anything
x=tf.compat.v1.placeholder(tf.float32)  # input 
```

```python
linear_model=w*x + b

init = tf.compat.v1.global_variables_initializer()  # initalizing variables
```

```python
s=tf.compat.v1.Session() # setting up session for computation

s.run(init)
s.run(w)
s.run(b)
```

```python
print("predicted output is")
print(s.run(linear_model,{x:[1,2,3,4]})) # applying this formula w*x + b with initial value of weight and bias
```

```python
y=tf.compat.v1.placeholder(tf.float32)  # y is for holding desired output
squared_deltas=tf.compat.v1.square(linear_model -y)
loss = tf.compat.v1.reduce_sum(squared_deltas) # reduce_sum is used to sum all the squared errors and give us rmse
```

```python
print("loss or Error is ")
print(s.run(loss, {x:[1,2,3,4], y:[0,-1,-2,-3]}))   #given input and output values here because we have used placeholder
```

```python
# now we want to reduce the rmse so we will use gradient descent optimizer

optimizer =  tf.compat.v1.train.GradientDescentOptimizer(0.01) # 0.01 is learning rate
train = optimizer.minimize(loss)
s.run(init)
for i in range(1000):  # 1000 is training epoch
 s.run(train, {x:[1,2,3,4,],y:[0,-1,-2,-3]})
```

```python
# Dataset Used : Sonar Dataset | Algorithm Used: Perceptron
import matplotlib.pyplot as plt
import tensorflow as tf
import numpy as np
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.utils import shuffle
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder , OneHotEncoder
 
tf.compat.v1.disable_eager_execution()
```

```python
import os
flower_types = os.listdir('G:\\Python coding\\Metropolia\\Computer Vision\\flowers')
flower_types
```

```python
import cv2
path = 'G:\\Python coding\\Metropolia\\Computer Vision\\flowers\\'

im_size = 60

images = []
labels = []

for i in flower_types:
    data_path = path + str(i)  # entered in rose folder
    filenames = [i for i in os.listdir(data_path) if i.endswith('.jpg')]
    #print(filenames)  # will get the names of all images which ends with .jpg extension
    for f in filenames:
        img = cv2.imread(data_path + '/' + f)  # reading that image as array
        #print(img)  # will get the image as an array
        img = cv2.resize(img, (im_size, im_size))
        images.append(img)
        labels.append(i)
```

```python
images = np.array(images)
images = images.astype('float32') / 255 
#images
```

```python
from sklearn.preprocessing import LabelEncoder , OneHotEncoder
x=images
y=labels
y_labelencoder = LabelEncoder ()
y = y_labelencoder.fit_transform (y)
print (y)
```

```python
y=y.reshape(-1,1)
#onehotencoder = OneHotEncoder(categorical_features=[0])  
#Y= onehotencoder.fit_transform(y)
#Y=Y.toarray()

from sklearn.compose import ColumnTransformer
ct = ColumnTransformer([('my_ohe', OneHotEncoder(), [0])], remainder='passthrough')
Y = ct.fit_transform(y) #.toarray()
print(Y)
```

```python
from sklearn.utils import shuffle
from sklearn.model_selection import train_test_split
x,Y = shuffle(x,Y,random_state=1)
train_x, test_x, train_y, test_y = train_test_split(images, Y, test_size=0.20, random_state=415)
```

```python
print(train_x.shape)
print(train_y.shape)
print(test_x.shape)
print(test_y.shape)
```

```python
#train_x=train_x.reshape(60*60*3)
#train_x.shape
train_x=np.reshape(train_x,(34,10800))
print(train_x.shape)

test_x=np.reshape(test_x,(9,10800))
test_x.shape
```

```python
learning_rate = 0.01
training_epochs = 1000
n_dim=60
n_class=2


x = tf.compat.v1.placeholder(tf.float32, [None, 10800])
W = tf.Variable(tf.zeros([10800,n_class]))
b = tf.Variable(tf.zeros([n_class]))
y_ = tf.compat.v1.placeholder(tf.float32,[None,n_class]) # actual y
```

```python
init = tf.compat.v1.global_variables_initializer()
```

```python
pred = tf.nn.softmax(tf.matmul(x, W)+ b) # perceptron
 
cost_function = tf.reduce_mean(tf.nn.softmax_cross_entropy_with_logits(logits=pred, labels=y_)) 
training_step = tf.compat.v1.train.GradientDescentOptimizer(learning_rate).minimize(cost_function)
```

```python
sess = tf.compat.v1.Session()
sess.run(init)
cost_history = []
```

```python
for epoch in range(training_epochs):
    sess.run(training_step,feed_dict={x:train_x,y_:train_y})
    cost = sess.run(cost_function,feed_dict={x: train_x,y_: train_y})
    cost_history = np.append(cost_history,cost)
    print('epoch : ', epoch,  ' - ', 'cost: ', cost)
 
pred_y = sess.run(pred, feed_dict={x: test_x})
```
